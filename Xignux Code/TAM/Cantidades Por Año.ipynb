```
# traer la tabla maestra de la base de datos
tabla = 'DataSet.masterv3'
master = spark.sql(f""" select * from {tabla} """)
```

```
# convertirlo en un dataframe de pandas
masterPD = master.toPandas()
```

```
len(masterPD)
```

```
# filtro con los requisitos del mercado 'MX PAD'
MXPAD = masterPD[masterPD['DESP_M'] == 'Residenciales']
MXPAD = MXPAD[MXPAD['CFCG_C'].isin(['FVA', 'PAR', 'FSA', 'FIG', 'CFE'])]
MXPAD = MXPAD[MXPAD['CITG_LOV'] == '1PM']
MXPAD
```

```
# por alguna razón hay más filas que en el otro query
len(MXPAD)
```

```
# crear columna con el año de la factura
import pandas as pd
MXPAD['INVD_F'] = MXPAD['INVD_F'].replace('0000-00-00', '1950-01-01')
MXPAD['INVD_F'] = pd.to_datetime(MXPAD['INVD_F'], format= '%Y-%d-%m')
MXPAD['year'] = pd.DatetimeIndex(MXPAD['INVD_F']).year
```

```
# ordered quantity, delivered quantity y recommended quantity agrupadas por año y sumadas
MXPAD[['OQUA_LOV', 'DQUA_LOV', 'RECQ_F', 'year']].groupby(['year']).sum()
```
