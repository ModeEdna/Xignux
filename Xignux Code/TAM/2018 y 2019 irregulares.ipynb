```
# podría ser que las diferencias en las cantidades pedidas venga de un uso de filtros diferentes a los de los otros años
```

```
# agarrar la tabla
tabla = 'DataSet.masterLimpia2'
master = spark.sql(f""" select * from {tabla} """)
```

```
# convertirla en un dataframe de pandas
A = master.toPandas()
```

```
# ver el dataframe
A
```

```
# crear una columna del año
import pandas as pd
A['year'] = pd.DatetimeIndex(A['INVD_F']).year
```

```
# filtrar por los años 2018 y 2019 y guardar la tabla a A
A = A[A['year'].between(2018, 2019)]
```

```
# filtrar por el mercado de Mex PD
A2 = A[A['DESP_M'] == 'Residenciales']
A2 = A2[A2['CFCG_C'].isin(['FVA', 'PAR', 'CFE'])]
A2 = A2[A2['CITG_LOV'] == '1PM']
A2
```

```
# ver la cantidad de filas en el año 2018 y 2019
A3 = A2[A2['year']==2018][['DQUA_LOV','OQUA_LOV','RECQ_F']]
A3['diff'] = A3['DQUA_LOV'] - A3['OQUA_LOV']
A3[A3['diff']>0]
```

```
# ver la cantidad de filas en el año 2019
A4 = A2[A2['year']==2019][['DQUA_LOV','OQUA_LOV','RECQ_F']]
A4['diff'] = A4['DQUA_LOV'] - A4['OQUA_LOV']
A4[A4['diff']>0]
```

```
# ver números agrupados por el año
A2.groupby(['year']).sum()
```

```
# cantidad de filas
len(A2)
```

```
# crear columna de diferencia entre cantidad pedida y cantidad entregada
A2['diff'] = A2['OQUA_LOV'] - A2['DQUA_LOV']
```

```
# ver las líneas con mayor entrega que pedido
A2.sort_values(['diff'], ascending=True)
```

```
# argupar valores por año y sumarlos
A2[['OQUA_LOV', 'RECQ_F', 'year', 'DQUA_LOV', 'diff']].groupby(['year']).sum()
```
