```
%sql

CREATE OR REPLACE TEMPORARY VIEW datos_odv AS
  SELECT 
      odv.orno                         orden_venta,
      trim(fam.desp)                   planta,
      trim(cte.cfcg)                   grupo_contable,
      trim(item.citg)                  familia,
      trim(odvc.chan)                  chn,
      trim(odvc.prop)                  mercado
  FROM 
    TTDSLS041006                       AS lin
    INNER JOIN TTDSLS040006            AS odv
        ON  lin.orno = odv.orno
    LEFT JOIN TTDPGE940006             AS odvc 
      ON odv.orno = odvc.orno
    INNER JOIN ttccom010007            AS cte
        ON odv.cuno = cte.cuno
    INNER JOIN ttiitm001006            AS item
        ON lin.item = item.item
    LEFT JOIN ttdpge808006             AS fam
        ON item.citg = fam.citg
```

```
%sql

CREATE OR REPLACE TEMPORARY VIEW tabla_mkt AS
SELECT
    orden_venta,
    planta,
    grupo_contable,
    familia,
    chn,
    mercado,
    CASE
    -- Privado
    WHEN planta = 'Residenciales'
        AND grupo_contable IN ('FSA', 'FVA', 'PAR', 'FIG')
        AND familia IN ('1PT', '2PT', '3PT')
        THEN 'Privado'
    -- CFE
    WHEN planta = 'Residenciales'
        AND grupo_contable = 'CFE'
        THEN 'CFE'
    -- Mx Pad
    WHEN planta = 'Residenciales'
        AND grupo_contable IN ('FSA', 'FVA', 'PAR', 'FIG')
        AND familia = '1PM'
        THEN 'Mx Pad'
    -- Sumergible
    WHEN planta = 'Residenciales'
        AND grupo_contable IN ('FSA', 'FVA', 'PAR', 'FIG')
        AND familia = '1SM'
        THEN 'Sumergible'
    -- LA Prolec
    WHEN planta = 'Residenciales'
        AND grupo_contable IN ('COL', 'EXP')
        AND familia IN ('1PT', '2PT', '3PT')
        THEN 'LA Prolec'
    -- Utility
    WHEN planta = 'Residenciales'
        AND grupo_contable = 'EX1'
        AND familia IN ('1PT', '2PT', '3PT')
        AND chn = 1
        AND SUBSTR(mercado,0,1) != 'U'
        THEN 'Utility'
    -- R/M
    WHEN planta = 'Residenciales'
        AND grupo_contable = 'EX1'
        AND familia IN ('1PT', '2PT', '3PT')
        AND chn = 1
        AND SUBSTR(mercado,0,1) = 'U'
        THEN 'R/M'
    -- C&I
    WHEN planta = 'Residenciales'
        AND grupo_contable = 'EX1'
        AND familia IN ('1PT', '2PT', '3PT')
        AND chn = 8
        THEN 'C' || chr(38) || 'I'
    -- INTERNACIONAL
    WHEN planta = 'Residenciales'
        AND grupo_contable = 'EX1'
        AND familia IN ('1PT', '2PT', '3PT')
        AND chn = 4
        THEN 'INTERNACIONAL'
    -- US Pad
    WHEN planta = 'Residenciales'
        AND grupo_contable = 'EX1'
        AND familia = '1PM'
        THEN 'US Pad'


    -- Utility
    WHEN planta = 'Comerciales'
        AND grupo_contable = 'EX1'
        AND familia = '3PP'
        AND chn = 1
        AND SUBSTR(mercado,0,1) != 'U'
        THEN 'Utility'
    -- R/M
    WHEN planta = 'Comerciales'
        AND grupo_contable = 'EX1'
        AND familia = '3PP'
        AND chn = 1
        AND SUBSTR(mercado,0,1) = 'U'
        THEN 'R/M'
    -- INTERNACIONAL
    WHEN planta = 'Comerciales'
        AND grupo_contable = 'EX1'
        AND familia = '3PP'
        AND chn = 4
        THEN 'INTERNACIONAL'
    -- Mx < 150
    WHEN planta = 'Comerciales'
        AND grupo_contable IN ('FSA', 'FVA', 'PAR', 'FIG')
        AND familia = '2PD'
        THEN 'Mx < 150'
    -- Mx > 150
    WHEN planta = 'Comerciales'
        AND grupo_contable IN ('FSA', 'FVA', 'PAR', 'FIG')
        AND familia = '2PS'
        THEN 'Mx > 150'
    -- Estación
    WHEN planta = 'Comerciales'
        AND grupo_contable IN ('FSA', 'FVA', 'PAR', 'FIG')
        AND familia = '2ES'
        THEN 'Estación'            
    ELSE
        NULL
    END AS mkt
FROM 
    datos_odv
```

```
%sql

SELECT
lin.ORNO AS Orden, lin.CUNO AS Cliente, lin.ITEM, lin.PRIC AS Precio, lin.OQUA AS Cantidad, lin.ODAT AS Fecha, item.DSCA, fam.CITG, fam.DESF, fam.DESP, item.SEAK, item.SEAB, mkt.mkt AS Mercado
FROM 
    TTDSLS041006                           AS lin
    INNER JOIN TTDSLS040006                AS odv
      ON  lin.orno = odv.orno
    INNER JOIN ttiitm001006                AS item
        ON lin.item = item.item
    LEFT JOIN ttdpge808006                AS fam
        ON item.citg = fam.citg
    INNER JOIN tabla_mkt                   AS mkt
        ON odv.orno = mkt.orden_venta
WHERE lin.ODAT >= '2016-01-01' AND trim(fam.CITG) = '1PM';
```
